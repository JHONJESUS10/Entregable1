```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generamos 50 tamaños de casas (en m²)
X = np.linspace(50, 250, 50)  # de 50 a 250 m²

# Precios simulados en miles de dólares, con algo de ruido
Y = 100 + 2.8 * X + np.random.randn(50) * 15
```

```python
modelo = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])
```

```python
modelo.compile(optimizer='adam', loss='mean_squared_error')

historial = modelo.fit(X, Y, epochs=200, verbose=0)
print("✅ ¡Modelo entrenado correctamente!")
```

```python
plt.figure(figsize=(6,4))
plt.plot(historial.history['loss'])
plt.title('Pérdida durante el entrenamiento del modelo')
plt.xlabel('Épocas')
plt.ylabel('Error cuadrático medio (MSE)')
plt.grid(True)
plt.show()
```

```python
# Pedimos al usuario que ingrese un tamaño
tamano = float(input("Ingrese el tamaño de la casa en m²: "))

# Convertimos el valor a un arreglo de NumPy y predecimos
precio_predicho = modelo.predict(np.array([[tamano]]))

print(f"\n📏 Tamaño ingresado: {tamano} m²")
print(f"💰 Precio estimado: ${precio_predicho[0][0]:.2f} mil dólares")
```

```python
# Convertimos X a una matriz columna (50, 1)
X_reshape = X.reshape(-1, 1)

# Gráfico con datos reales, línea de regresión y predicción ingresada
plt.figure(figsize=(7,5))
plt.scatter(X, Y, color='blue', label='Datos reales')
plt.plot(X, modelo.predict(X_reshape), color='red', label='Línea de regresión')
plt.scatter(tamano, precio_predicho, color='green', s=100, label='Predicción ingresada')

plt.title('Predicción del precio de casas según tamaño')
plt.xlabel('Tamaño (m²)')
plt.ylabel('Precio (miles de $)')
plt.legend()
plt.grid(True)
plt.show()
```

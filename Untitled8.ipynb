```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generamos 50 tama√±os de casas (en m¬≤)
X = np.linspace(50, 250, 50)  # de 50 a 250 m¬≤

# Precios simulados en miles de d√≥lares, con algo de ruido
Y = 100 + 2.8 * X + np.random.randn(50) * 15
```

```python
modelo = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])
```

```python
modelo.compile(optimizer='adam', loss='mean_squared_error')

historial = modelo.fit(X, Y, epochs=200, verbose=0)
print("‚úÖ ¬°Modelo entrenado correctamente!")
```

```python
plt.figure(figsize=(6,4))
plt.plot(historial.history['loss'])
plt.title('P√©rdida durante el entrenamiento del modelo')
plt.xlabel('√âpocas')
plt.ylabel('Error cuadr√°tico medio (MSE)')
plt.grid(True)
plt.show()
```

```python
# Pedimos al usuario que ingrese un tama√±o
tamano = float(input("Ingrese el tama√±o de la casa en m¬≤: "))

# Convertimos el valor a un arreglo de NumPy y predecimos
precio_predicho = modelo.predict(np.array([[tamano]]))

print(f"\nüìè Tama√±o ingresado: {tamano} m¬≤")
print(f"üí∞ Precio estimado: ${precio_predicho[0][0]:.2f} mil d√≥lares")
```

```python
# Convertimos X a una matriz columna (50, 1)
X_reshape = X.reshape(-1, 1)

# Gr√°fico con datos reales, l√≠nea de regresi√≥n y predicci√≥n ingresada
plt.figure(figsize=(7,5))
plt.scatter(X, Y, color='blue', label='Datos reales')
plt.plot(X, modelo.predict(X_reshape), color='red', label='L√≠nea de regresi√≥n')
plt.scatter(tamano, precio_predicho, color='green', s=100, label='Predicci√≥n ingresada')

plt.title('Predicci√≥n del precio de casas seg√∫n tama√±o')
plt.xlabel('Tama√±o (m¬≤)')
plt.ylabel('Precio (miles de $)')
plt.legend()
plt.grid(True)
plt.show()
```
